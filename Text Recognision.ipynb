```python


import cv2
import pytesseract
import matplotlib.pyplot as plt

# Set the path to the Tesseract OCR executable
pytesseract.pytesseract.tesseract_cmd = r"E:\vsc\tesseract.exe"
img = cv2.imread(r"C:\Users\zbook 17 g3\Desktop\New folder (4)\TEXTO.png")
hi,w,_=img.shape
#recognize
rects=pytesseract.image_to_boxes(img)
#print(rects)
font = cv2.FONT_HERSHEY_SIMPLEX
font_scale = 1.5
font_thickness = 2
font_color = (198, 98, 155)
#drawing rectangles
for rect in rects.splitlines():
    rect=rect.split(" ")
    print(rect)
    x,y,w,h=int(rect[1]),int(rect[2]),int(rect[3]),int(rect[4])
    cv2.rectangle(img,(x,hi-y),(w,hi-h),(155,186,99),5)
    cv2.putText(img,rect[0],(x,hi-y+26),font, font_scale, font_color, font_thickness)
    
    
#cv2.imshow("image",img) 
#cv2.waitKey(0)
#cv2.destroyAllWindows()   
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.title("Text")  
plt.axis("off")  

    

```
